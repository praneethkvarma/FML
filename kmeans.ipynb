```python
from sklearn.cluster import KMeans
import numpy as np 
```

```python
x=np.array([[1,2],[1,4],[1,0],[10,2],[10,4],[10,0]])
kmeans=KMeans(n_clusters=2 ,random_state=0).fit(x)
```

```python
kmeans.labels_
```

```python
kmeans.predict([[0,0],[12,3]])
```

```python
kmeans.cluster_centers_
```

```python
from sklearn.cluster import KMeans 
from sklearn import preprocessing 
import sklearn.cluster as cluster 
import sklearn.metrics as metrics  
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns 
from matplotlib import pyplot as plt 
%matplotlib inline
```

```python
df=pd.read_csv("/content/Mall_Customers.csv")
df.head()
```

```python
df.shape
```

```python
scaler=MinMaxScaler()
scale=scaler.fit_transform(df[['Annual Income (k$)','Spending Score (1-100)']])
df_scale = pd.DataFrame(scale, columns = ['Annual Income (k$)','Spending Score (1-100)']);
df_scale.head(5)
```

```python
km=KMeans(n_clusters=2)
y_predicted = km.fit_predict(df_scale[['Annual Income (k$)','Spending Score (1-100)']])
y_predicted
```

```python
km.cluster_centers_
```

```python
df['Clusters']=km.labels_ 
sns.scatterplot(x="Spending Score (1-100)",y="Annual Income (k$)",hue='Clusters',data=df,palette='viridis')
```

```python
km=KMeans(n_clusters=5)
y_predicted=km.fit_predict(df_scale[['Annual Income (k$)','Spending Score (1-100)']])
y_predicted
```

```python
df['Clusters']=km.labels_ 
sns.scatterplot(x='Annual Income (k$)',y='Spending Score (1-100)',hue='Clusters',data=df)
```
